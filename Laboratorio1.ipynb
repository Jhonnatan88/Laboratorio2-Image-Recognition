```python
def funcion_prblema3 (path,color):

    img = cv2.imread(path)
    
    alto  = img.shape[0]
    ancho = img.shape[1]
    lienzo = np.zeros((alto, ancho, 3))

    for i in range(0,alto):
        for j in range(0,ancho):
            pixel = img[i,j]
            blue  = pixel[2]
            green = pixel[1]
            red   = pixel[0]
#en esta libreria es azul, verde y rojo
            if color == 'azul':
                lienzo[i,j] = [blue,0,0]
            elif color == 'rojo':
                lienzo[i,j] = [0,0,red]
            elif color == 'verde':
                lienzo[i,j] = [0,green,0]
    print(color)       
    nombre = str(color+"_lienzo.jpg")
    cv2.imwrite(nombre, lienzo)
    
   
    

def imagenes_grises (nombre):
    nombre1 = str(nombre+"_lienzo.jpg")
    img2 = cv2.imread(nombre1)
   
    alto_resize = img2.shape[0]
    ancho_resize = img2.shape[1]
    media_ponderada_a = np.zeros((alto_resize, ancho_resize, 3))

    for i in range(0, alto_resize):
        for j in range(0, ancho_resize):
            pixel_a = img2[i, j]
            
            blue_a  = pixel_a[2]
            green_a  = pixel_a[1]
            red_a  = pixel_a[0]
           
            if nombre == 'azul':
                media_ponderada_a[i, j] = int(blue_a + 0 + 0)
            elif nombre == 'rojo':
                media_ponderada_a[i, j] = int(0 + 0 +red_a)
            elif nombre == 'verde':    
                media_ponderada_a[i, j] = int(0 + green_a +0)
                
    cv2.imwrite('gris_'+nombre+'.jpg',media_ponderada_a)
    img = cv2.imread('gris_'+nombre+'.jpg')
    alto = img.shape[0]
    ancho = img.shape[1]
    media_ponderada_b = np.zeros((alto, ancho, 3))

    
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            
            blue  = pixel[2]
            green = pixel[1]
            red   = pixel[0]
           
            #en esta libreria es azul, verde y rojo
            media_ponderada_b[i,j] = int(0.299*blue + 0.587*green + 0.114*red)

    cv2.imwrite('gris_'+nombre+'.jpg',media_ponderada_b)       
    img2= cv2.imread('gris_'+nombre+'.jpg')
    img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show() 

def imagenes_escala_gris():
    path1 = str(input("Ingrese la ruta del archvio de la imagen "))
    color1 = str(input("Ingrese el nombre de color a convertir "))
    path2 = str(input("Ingrese la ruta del archvio de la imagen "))
    color2 = str(input("Ingrese el nombre de color a convertir "))
    path3 = str(input("Ingrese la ruta del archvio de la imagen "))
    color3 = str(input("Ingrese el nombre de color a convertir "))

    #print(path1,color1)
    funcion_prblema3(path1, color1)
    funcion_prblema3(path2, color2)
    funcion_prblema3(path3, color3)
    
    imagenes_grises(color1)
    imagenes_grises(color2)
    imagenes_grises(color3)
```

Investigue brevemente en que consiste el espacio de color HSV y como se mapean colores a dicho
espacio, para entregar este ejercicio puede hacerlo por medio de Markdown en el mismo Notebook

El espacio de color es una forma de representar los canales de color presentes en una imagen. Hay varios diferentes espacios de color, y cada uno tiene su propio significado.

Algunos espacios de color comunes son RGB (rojo, verde, azul), HSV (croma, saturación, valor). 

Espacio de color BGR: el espacio de color predeterminado de OpenCV es RGB. Sin embargo, en realidad almacena el color en formato BGR. Es un modelo de color aditivo donde las diferentes intensidades de Azul, Verde y Rojo dan distintas tonalidades de color.

Espacio de color HSV: almacena información de color en una representación cilíndrica de puntos de color RGB. Intenta representar los colores tal como los percibe el ojo humano. El valor de tono varía de 0 a 179, el valor de saturación varía de 0 a 255 y el valor de valor varía de 0 a 255. Se utiliza principalmente con fines de segmentación de color.

Espacio de color CMYK: a diferencia de RGB, es un espacio de color sustractivo. El modelo CMYK funciona enmascarando parcial o totalmente los colores sobre un fondo más claro, generalmente blanco. La tinta reduce la luz que de otro modo se reflejaría. Este modelo se llama sustractivo porque las tintas "restan" los colores rojo, verde y azul de la luz blanca. La luz blanca menos las hojas rojas cian, la luz blanca menos las hojas verdes magenta y la luz blanca menos las hojas azules son amarillas.
